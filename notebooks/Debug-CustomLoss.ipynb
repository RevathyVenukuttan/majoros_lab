```python
import tensorflow as tf
from tensorflow.keras.models import model_from_json
from keras_nlp.layers import RotaryEmbedding, TransformerEncoder
# from keras.layers import TFOpLambda

# File paths
json_path = '/hpc/group/igvf/revathy/models/K562/TE/K562_TE-6.json'
weights_path = '/hpc/group/igvf/revathy/models/K562/TE/K562_TE-6.h5'

# Load model architecture
with open(json_path, "r") as f:
    model_json = f.read()

custom_objects = {
    "RotaryEmbedding": RotaryEmbedding,
    "TransformerEncoder": TransformerEncoder,
    # "TFOpLambda": tf.keras.layers.TFOpLambda  # this may resolve correctly via tf.keras
}

# Reconstruct model
model = model_from_json(model_json, custom_objects=custom_objects)

# Load weights
model.load_weights(weights_path)

# Display summary
model.summary()

```
