```python
import gzip
import re
import subprocess
import os
import shutil
```

```python

input_fasta = "/hpc/home/rv103/igvf/K562/full-set/300bp/data-normalized/test.fasta.gz"          # Input 300bp FASTA
genome_2bit = "/hpc/home/rv103/igvf/hg38/hg38.2bit"                  # .2bit genome file
temp_bed = "/hpc/home/rv103/igvf/revathy/sequences/temp/temp_regions.bed"             # Temporary BED file
temp_fasta = "/hpc/home/rv103/igvf/revathy/sequences/temp/temp_1000bp.fa"             # Temporary uncompressed FASTA
output_fasta_gz = "/hpc/home/rv103/igvf/revathy/sequences/K562/1000bp/test.fasta.gz"   # Final compressed FASTA

with gzip.open(input_fasta, "rt") as infile, open(temp_bed, "w") as bed_out:
    for line in infile:
        if line.startswith(">"):
            match = re.search(r"/coord=(chr[\w\d]+):(\d+)-(\d+)", line)
            if not match:
                continue
            chrom, start, end = match.groups()
            start = int(start)
            end = int(end)
            ext_start = max(0, start - 350)
            ext_end = end + 350
            name = f"{chrom}:{ext_start}-{ext_end}"
            bed_out.write(f"{chrom}\t{ext_start}\t{ext_end}\t{name}\n")

# Step 2: Run twobittofa to extract sequences
cmd = ["/hpc/group/igvf/revathy/software/twoBitToFa", f"-bed={temp_bed}", genome_2bit, temp_fasta]
print("Running:", " ".join(cmd))
subprocess.run(cmd, check=True)

# Step 3: Compress the resulting FASTA file
with open(temp_fasta, 'rb') as f_in, gzip.open(output_fasta_gz, 'wb') as f_out:
    shutil.copyfileobj(f_in, f_out)

# Step 4: Cleanup temporary files
os.remove(temp_bed)
os.remove(temp_fasta)

print(f"Compressed FASTA written to: {output_fasta_gz}")
```

#### Check how many sequences in each set has length not equal to 1000bp

```python
from Bio import SeqIO
import gzip
import os

# List of input FASTA files
fasta_files = {
    "Train": "/hpc/home/rv103/igvf/revathy/sequences/K562/1000bp/train.fasta.gz",
    "Validation": "/hpc/home/rv103/igvf/revathy/sequences/K562/1000bp/validation.fasta.gz",
    "Test": "/hpc/home/rv103/igvf/revathy/sequences/K562/1000bp/test.fasta.gz"
}

expected_length = 1000

for label, filepath in fasta_files.items():
    if not os.path.exists(filepath):
        print(f"{label} file not found: {filepath}")
        continue

    print(f"\n Checking {label} file: {os.path.basename(filepath)}")
    count = 0
    bad_seqs = []

    with gzip.open(filepath, "rt") as handle:
        for idx, record in enumerate(SeqIO.parse(handle, "fasta")):
            count += 1
            if len(record.seq) != expected_length:
                bad_seqs.append((idx + 1, len(record.seq)))

    print(f"Total sequences: {count}")
    if bad_seqs:
        print(f"{len(bad_seqs)} sequences are not {expected_length} bp long:")
        for seq_idx, seq_len in bad_seqs:
            print(f"  - Sequence {seq_idx}: {seq_len} bp")
    else:
        print(f"All sequences are exactly {expected_length} bp long.")
```

#### Updated script that filters for sequences that are not 1000bp

```python
import gzip
import re
import subprocess
import os
from Bio import SeqIO
from Bio.Seq import Seq

input_fasta = "/hpc/home/rv103/igvf/K562/full-set/300bp/data-normalized/train.fasta.gz"  # Input 300bp FASTA
genome_2bit = "/hpc/home/rv103/igvf/hg38/hg38.2bit"                                       # .2bit genome file
temp_bed = "/hpc/home/rv103/igvf/revathy/sequences/temp/temp_regions.bed"                # Temporary BED file
temp_fasta = "/hpc/home/rv103/igvf/revathy/sequences/temp/temp_1000bp.fa"                # Temporary uncompressed FASTA
output_fasta_gz = "/hpc/home/rv103/igvf/revathy/sequences/K562/1000bp/train.fasta.gz"    # Final compressed FASTA
expected_length = 1000

# Step 1: Generate BED file from 300bp input FASTA headers
with gzip.open(input_fasta, "rt") as infile, open(temp_bed, "w") as bed_out:
    for line in infile:
        if line.startswith(">"):
            match = re.search(r"/coord=(chr[\w\d]+):(\d+)-(\d+)", line)
            if not match:
                continue
            chrom, start, end = match.groups()
            start = int(start)
            end = int(end)
            ext_start = max(0, start - 350)
            ext_end = end + 350
            name = f"{chrom}:{ext_start}-{ext_end}"
            bed_out.write(f"{chrom}\t{ext_start}\t{ext_end}\t{name}\n")

# Step 2: Run twoBitToFa to extract sequences
cmd = ["/hpc/group/igvf/revathy/software/twoBitToFa", f"-bed={temp_bed}", genome_2bit, temp_fasta]
print("Running:", " ".join(cmd))
subprocess.run(cmd, check=True)

# Step 3: Convert sequences to uppercase and keep only 1000bp entries
valid_count = 0
removed_count = 0

with gzip.open(output_fasta_gz, 'wt') as f_out:
    with open(temp_fasta, 'r') as f_in:
        for record in SeqIO.parse(f_in, "fasta"):
            if len(record.seq) == expected_length:
                record.seq = Seq(str(record.seq).upper())
                SeqIO.write(record, f_out, "fasta")
                valid_count += 1
            else:
                removed_count += 1

# Step 4: Cleanup temporary files
os.remove(temp_bed)
os.remove(temp_fasta)

# Final log
print(f"Compressed FASTA written to: {output_fasta_gz}")
print(f"Kept {valid_count} sequences; removed {removed_count} sequences not {expected_length}bp long.")
```

### Print sequences that are not thousand base pairs

```python
import gzip
import re
import subprocess
import os
from Bio import SeqIO
from Bio.Seq import Seq

genome_2bit = "/hpc/home/rv103/igvf/hg38/hg38.2bit"   # .2bit genome file
twoBitToFa = "/hpc/group/igvf/revathy/software/twoBitToFa"
expected_length = 1000

base_in = "/hpc/home/rv103/igvf/K562/full-set/300bp/data-normalized"
base_out = "/hpc/home/rv103/igvf/revathy/sequences/K562/1000bp"
temp_dir = "/hpc/home/rv103/igvf/revathy/sequences/temp"

os.makedirs(base_out, exist_ok=True)
os.makedirs(temp_dir, exist_ok=True)


def process_set(setname):
    input_fasta = f"{base_in}/{setname}.fasta.gz"
    temp_bed = f"{temp_dir}/{setname}.bed"
    temp_fasta = f"{temp_dir}/{setname}.fa"
    output_fasta_gz = f"{base_out}/{setname}.fasta.gz"

    # Step 1: Generate BED file from FASTA headers
    with gzip.open(input_fasta, "rt") as infile, open(temp_bed, "w") as bed_out:
        for line in infile:
            if line.startswith(">"):
                match = re.search(r"/coord=(chr[\w\d]+):(\d+)-(\d+)", line)
                if not match:
                    continue
                chrom, start, end = match.groups()
                start, end = int(start), int(end)
                ext_start = max(0, start - 350)
                ext_end = end + 350
                name = f"{chrom}:{ext_start}-{ext_end}"
                bed_out.write(f"{chrom}\t{ext_start}\t{ext_end}\t{name}\n")

    # Step 2: Run twoBitToFa
    cmd = [twoBitToFa, f"-bed={temp_bed}", genome_2bit, temp_fasta]
    print(f"[{setname}] Running:", " ".join(cmd))
    subprocess.run(cmd, check=True)

    # Step 3: Filter sequences to 1000bp, report non-1000bp
    valid_count, removed_count = 0, 0
    bad_headers = []

    with gzip.open(output_fasta_gz, "wt") as f_out, open(temp_fasta, "r") as f_in:
        for record in SeqIO.parse(f_in, "fasta"):
            seqlen = len(record.seq)
            if seqlen == expected_length:
                record.seq = Seq(str(record.seq).upper())
                SeqIO.write(record, f_out, "fasta")
                valid_count += 1
            else:
                removed_count += 1
                header = record.description if record.description else record.id
                print(f"[{setname}] NON-{expected_length}bp: {header}\tlen={seqlen}")
                bad_headers.append((header, seqlen))

    # Step 4: Cleanup temps
    for f in (temp_bed, temp_fasta):
        try:
            os.remove(f)
        except FileNotFoundError:
            pass

    # Final log
    print(f"[{setname}] Compressed FASTA written to: {output_fasta_gz}")
    print(f"[{setname}] Kept {valid_count} sequences; removed {removed_count} not {expected_length}bp.")
    if bad_headers:
        print(f"[{setname}] Summary of non-{expected_length}bp headers:")
        for header, seqlen in bad_headers[:20]:
            print(f"   {header}\tlen={seqlen}")
        if len(bad_headers) > 20:
            print(f"   ... and {len(bad_headers)-20} more.")


# Run for all three sets
for setname in ["train", "validation", "test"]:
    process_set(setname)
```
